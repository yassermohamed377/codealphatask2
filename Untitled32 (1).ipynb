```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
import yfinance as yf
```

```python
company = 'TSLA'
df = yf.download(company, start='2015-01-01', end='2023-01-01')
```

```python
df = df[['Close']]
df.head()
```

```python
plt.figure(figsize=(10,6))
plt.plot(df.index, df['Close'], label=f'{company} Closing Price')
plt.xlabel('Date')
plt.ylabel('Closing Price')
plt.title(f'{company} Stock Price')
plt.legend()
plt.show()
```

```python
scaler = MinMaxScaler(feature_range=(0,1))
scaled_data = scaler.fit_transform(df)
```

```python
train_size = int(len(scaled_data) * 0.8)
train_data = scaled_data[:train_size]
test_data = scaled_data[train_size:]
```

```python
def create_dataset(data, time_step=60):
    X, y = [], []
    for i in range(time_step, len(data)):
        X.append(data[i-time_step:i, 0])
        y.append(data[i, 0])
    return np.array(X), np.array(y)
```

```python
time_step = 60
X_train, y_train = create_dataset(train_data, time_step)
X_test, y_test = create_dataset(test_data, time_step)
```

```python
X_train = X_train.reshape((X_train.shape[0], X_train.shape[1], 1))
X_test = X_test.reshape((X_test.shape[0], X_test.shape[1], 1))
```

```python
model = Sequential()
model.add(LSTM(units=50, return_sequences=True, input_shape=(X_train.shape[1], 1)))
model.add(Dropout(0.2))
model.add(LSTM(units=50, return_sequences=False))
model.add(Dropout(0.2))
model.add(Dense(units=25))
model.add(Dense(units=1))
```

```python
# Compile the model
model.compile(optimizer='adam', loss='mean_squared_error')
```

```python
history = model.fit(X_train, y_train, batch_size=64, epochs=50, validation_data=(X_test, y_test))
```

```python
predictions = model.predict(X_test)
predictions = scaler.inverse_transform(predictions)
```

```python
plt.figure(figsize=(10,6))
plt.plot(df.index[train_size + time_step:], scaler.inverse_transform(test_data[time_step:]), label='Actual Price')
plt.plot(df.index[train_size + time_step:], predictions, label='Predicted Price')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.show()
```

```python
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(scaler.inverse_transform(test_data[time_step:]), predictions)
print(f'Mean Squared Error: {mse}')
```
